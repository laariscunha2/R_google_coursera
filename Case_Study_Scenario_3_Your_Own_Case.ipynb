##**STEP 1:**
Installing packages

```r
install.packages("tidyverse")
install.packages("ggplot2")

library(tidyverse)
library(ggplot2)
```

## **STEP 2:**
Importing dataset

```r
fish_df <- read.csv("/content/fish_catches2.csv")
head(fish_df)
```

## **STEP 3:**
transformation

```r
fish_df <- fish_df %>%
  mutate(Country = recode(Country,
                  " UK (Channel Island Guernsey)" = "UK (Guernsey)" ,
                  " UK (Channel Island Jersey)" = "UK (Jersey)",
                  "LY" = "Libya",
                  "CN" = "China",
                  "JP" = "Japan" ,
                  "TW" = "Taiwan"))
```

*i've transformed the names for shorter one's so the graphic can look better*

```r
unique_fish <- unique(fish_df$Species)
length(unique_fish)
```

## STEP 4:
Analysis

```r
ggplot(data=fish_df, mapping = aes(x=Country, y=year_2006)) +
  geom_smooth(stat = "identity", color = "darkblue", position = position_dodge(width = 2), size=3.5) +
  theme(axis.text.x = element_text(angle = 90)) +
  scale_y_continuous(labels = scales::comma) +
  labs(title="Fishing: Country vs. Quantity", subtitle = "2006")
```
